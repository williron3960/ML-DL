```python
import pandas as pd  
import numpy as np  
import seaborn as sns
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
```

```python
pw2014_ini=pd.read_csv('homeF_2014_merge_2.csv')
pw2015_ini=pd.read_csv('homeF_2015_merge_2.csv')
```

```python
power = pd.concat([pw2014_ini,pw2015_ini],axis=0,ignore_index=True)
```

```python
#descriptive statistics summary
target='Usage [kW]'
power[target].describe()
```

```python
#histogram
sns.distplot(power[target]);
```

```python
#skewness and kurtosis
print("Skewness: %f" % power[target].skew())
print("Kurtosis: %f" % power[target].kurt())
```

```python
#correlation matrix
corrmat = power.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=1, square=True);
```

```python
#Usage[kW] correlation matrix
k = 20 #number of variables for heatmap
cols = corrmat.nlargest(k, target)[target].index
cm = power[cols].corr()
```

```python
##Usage[kW] correlation matrix
sns.set(font_scale=1.25)
f, ax = plt.subplots(figsize=(16,12 ))
hm = sns.heatmap(cm,vmax=1, cbar=True, annot=True, square=True,linewidths=0, fmt='.2f', annot_kws={'size': 12}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
#ç”±ä¸Šåœ–å¯çŸ¥ èˆ‡ Usage[kW]æœ€å¤§é—œè¯çš„å…©å€‹åƒæ•¸ç‚º Phase_A,Phase_b

#æª¢æŸ¥ç¼ºå¤±å€¼
power.isnull().any()
```

```python
#ç§»é™¤ç¼ºå¤±å€¼
power = power.fillna(method='ffill')
```

```python
#å°‡è³‡æ–™åˆ†æˆå±¬æ€§åŠæ¨™ç±¤

X = power[['time','apparentTemperature','hr',]] #ä»¥æ™‚é–“,æº«åº¦,å°æ™‚ç‚ºç‰¹å¾µä¾†é æ¸¬ç”¨é›»é‡
Y = power['Usage [kW]']
```

```python
#å»ºç«‹è¨“ç·´é›†å’Œæ¸¬è©¦é›†

#2014å¹´1æœˆ~2015å¹´8æœˆç‚ºè¨“ç·´é›†,2015å¹´9æœˆ~2015å¹´12æœˆç‚ºæ¸¬è©¦é›†
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.167,shuffle=False)
#print('X_train: \n',X_train)
#print('X_test: \n',X_test)
#print('Y_train: \n',Y_train)
#print('Y_test: \n',Y_test)
```

```python
#å»ºç«‹ä¸¦è¨“ç·´æ¨¡åž‹

regressor = LinearRegression()  
regressor.fit(X_train, Y_train)
```

```python
#åˆ†æžè¶¨å‹¢ç·š

a  = regressor.intercept_#æˆªè·
b = regressor.coef_#è¿´æ­¸ä¿‚æ•¸
print("æ“¬åˆåƒæ•¸:æˆªè·",a,",è¿´æ­¸ä¿‚æ•¸ï¼š",b)
```

```python
#é¡¯ç¤ºç·šæ€§æ–¹ç¨‹ï¼Œä¸¦é™åˆ¶åƒæ•¸çš„å°æ•¸ä½çˆ²å…©ä½

print("æœ€ä½³æ“¬åˆç·š: Y = ",round(a,2),"+",round(b[0],2),"* X1 + ",round(b[1],2),"* X2",round(b[2],2),"* X3")
```

```python
Y_pred=regressor.predict(X_test)
df = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
df1 = df.head(25) #é¡¯ç¤ºå¯¦éš›å€¼å’Œé æ¸¬å€¼
df
```

```python
print(' ')
print('#è¦–è¦ºåŒ–')
df1.plot(kind='bar',figsize=(10,5))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
#å¹³å‡çµ•å°èª¤å·®(MAE)ã€å‡æ–¹èª¤å·®(MSE)ã€å‡æ–¹æ ¹èª¤å·®(RMSE)
print('*Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
print(' ')  
print('æª¢æŸ¥æº–ç¢ºåº¦(%): ',regressor.score(X_train, Y_train)*100)
```
